## Assignment 4: Tuition (6 points)

A college charges tuition according to the following:
- 12 or more credits: $20,000/semester

- 1-11 credits: $1,200 + $1,700 per credit

Write a program that prompts the user for a number of credits and determines and displays the tuition for the semester. As part of your program, define and use the following function:
```python
def calcTuition(credits)
```
The function returns the tuition for a given number of credits

````python
# Assignment 1

def getValues(n: int) -> list:
  '''
  Creates and returns list of weights
  Inputs n number of items into list
  Outputs a list
  '''
  items = []

  for i in range(n):
    try:
      item = float(input('Enter a weight value: '))
    except Exception as e:
      print(e)
      item = float('nan')
    items.append(item)


  return items

def calcAverage(alist: list) -> float:
  '''
  Returns the average of list
  Inputs a list of weights
  Outputs the average of the items
  '''
  try:
    return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
    print(e)
    return float('nan')

def main():
  num = int(input('How many weights? '))
  if num <= 0:
    raise ValueError('num must be positive')

  weights = getValues(num)
  avg = calcAverage(weights)
  print(f'Average weight: {avg} lbs.')

main()
````

````python
# Assignment 3

def getParkingHours():
  hours = int(input('How many hours parked? '))
  return hours

def calcParkingFee(h):
  cost = 6
  if h == 0:
    raise ValueError('Parking time must be positive')
  if h < 3:
    return cost
  elif h > 8:
    cost = 20
  else:
    cost = h * 2.5
  return cost

def main():
  hours = getParkingHours()
  fee = calcParkingFee(hours)
  print(f'Parking fee: ${fee:.2f}')

main()
````

````python
# Assignment 4

def calcTuition(credits):
  if credits <= 0:
    raise ValueError('Credits must be positive')
  elif credits >= 12:
    return 20000
  else:
    return 1200 + 1700 * credits

def main():
  credits = int(input('How many credits? '))
  tuition = calcTuition(credits)
  print(f'Tuition is ${tuition}')

main()
````
